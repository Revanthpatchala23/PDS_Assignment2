b) Remove the units from some of the attributes and only keep the numerical values (for
example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from
“New_price”).

```python
df['Mileage'] = df['Mileage'].str.replace(r'\s+[a-zA-Z/]+', '', regex=True)
df['Engine'] = df['Engine'].str.replace(r'\s+CC', '', regex=True)
df['Power'] = df['Power'].str.replace(r'\s+bhp', '', regex=True)
df['New_Price'] = df['New_Price'].str.replace(r'\s+Lakh', '', regex=True)

df.head()
```

```python
# replace the missing values in Mileage, Engine, Power and Seats columns
```

```python
# Clean and convert 'Mileage' to numeric
df['Mileage'] = df['Mileage'].str.extract('([\d\.]+)').astype(float)
df['Mileage'].fillna(df['Mileage'].median(), inplace=True)

# Clean and convert 'Engine' to numeric
df['Engine'] = df['Engine'].str.extract('([\d\.]+)').astype(float)
df['Engine'].fillna(df['Engine'].median(), inplace=True)

# Clean and convert 'Power' to numeric
df['Power'] = df['Power'].str.extract('([\d\.]+)').astype(float)
df['Power'].fillna(df['Power'].median(), inplace=True)

# Fill missing 'Seats' with mode
df['Seats'].fillna(df['Seats'].mode()[0], inplace=True)

```

```python
# dropping the 'New_Price' column since there are lot of missing values
```

```python
df.drop('New_Price', axis=1, inplace=True)
```

```python
print(df.isnull().sum())
```

c) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot
encoded value.

```python
fuel_mapping = {'Diesel': 0, 'Petrol': 1, 'Electric': 2}
df['Fuel_Type'] = df['Fuel_Type'].replace(fuel_mapping)

transmission_mapping = {'Manual': 0, 'Automatic': 1}
df['Transmission'] = df['Transmission'].replace(transmission_mapping)

df.head()
```

```python
# getting current year to calculate age of the car
```

```python
from datetime import datetime

current_year = datetime.now().year
```

d) Create one more feature and add this column to the dataset (you can use mutate function in
R for this). For example, you can calculate the current age of the car by subtracting “Year” value
from the current year.

```python
df['Age_of_Car'] = current_year - df['Year']
```

```python
df['Power_per_Liter'] = (df['Power'].astype(float) / df['Mileage'].astype(float)).round(2)
```

```python
df['Mileage_per_Seat'] = (df['Mileage'].astype(float) / df['Seats'].astype(float)).round(2)
```

Selecting columns

```python
# Select specific columns
selected_df = df[['Name', 'Year', 'Fuel_Type', 'Mileage', 'Transmission']]

# Filter rows based on a condition
filtered_df = df[df['Kilometers_Driven'] > 100000]

# Rename columns
renamed_df = df.rename(columns={'Name': 'Car_Name', 'Year': 'Manufacturing_Year'})

# Mutate/Create a new column
df['Price_Per_KM'] = df['Price'] / df['Kilometers_Driven']


# Arrange/Sort the dataframe by a column
arranged_df = df.sort_values('Age_of_Car', ascending=False)

# Summarize/Group by operations
grouped_df = df.groupby('Fuel_Type')['Kilometers_Driven'].mean()
grouped_df_2 = df.groupby('Transmission')['Price'].max()

print("Selected Columns:")
print(selected_df.head())
print("\nFiltered Dataframe:")
print(filtered_df.head())
print("\nRenamed Dataframe:")
print(renamed_df.head())
print("\nMutated Dataframe:")
print(df.head())
print("\nArranged Dataframe:")
print(arranged_df.head())
print("\nGrouped Dataframe (Average Kilometers Driven per Fuel Type):")
print(grouped_df)
print("\nGrouped Dataframe (Maximum New_Price per Transmission Type):")
grouped_df_2
df.to_csv('data_manipulation.csv', index=False)
```
